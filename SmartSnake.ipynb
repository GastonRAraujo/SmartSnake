```python
#EL JUEGO


#Ctes-------------------------------------------------------

WIDTH = 600
HEIGHT = 600

blue=(0,0,255)
red=(255,0,0)
green=(0,255,0)
white=(255,255,255)
black = (0, 0, 0)
 

snake_block = 10

speed = 20


Num_Acelerador = 0 #cant de comida para q vaya mas rapido
Acelerador = 0 #cantidad q acelera


LR = 1e-3
goal_steps = 300
score_requirement = 50
initial_games = 5000


#Matrices de rotacion------------------------------------------
R=np.matrix([[0,1], [-1,0]])

L=np.matrix([[0,-1], [1,0]])

I=np.identity(2)



#Clases---------------------------------------------------------

class Vibora():
  
      PosX = WIDTH / 2
  
      PosY = HEIGHT / 2
  
      snake_x = [PosX,PosX,PosX]
      snake_y = [PosY,PosY+10,PosY+20]
        
      
      def move(self, Action, vector, PosX, PosY):
        
        self.snake_x[0] = PosX
        self.snake_y[0] = PosY
    
            
        if Action == 1:
            Rot=np.matrix([[0,1], [-1,0]])
        elif Action == 2:
            Rot=np.matrix([[0,-1], [1,0]])
        elif Action == 0:
            Rot=np.identity(2)
        
        vector = np.dot(Rot,vector)
        
        Right = np.matrix([[1],[0]])
        Left = np.matrix([[-1],[0]])
        Up = np.matrix([[0],[1]])
        Down = np.matrix([[0],[-1]])
        
        if np.array_equal(vector,Left):
             self.PosX += -10
                    
        elif np.array_equal(vector,Right):
             self.PosX += 10
                    
        elif np.array_equal(vector,Up):
             self.PosY += -10
                    
        elif np.array_equal(vector,Down):
             self.PosY += 10

    
        for i in range(1,len(self.snake_x)):
            j = len(self.snake_x) - i
            self.snake_x[j] = self.snake_x[j-1]
        for i in range(1,len(self.snake_y)):
            j = len(self.snake_y) - i
            self.snake_y[j] = self.snake_y[j-1]

        self.snake_x[0] = self.PosX
        self.snake_y[0] = self.PosY
        self.Vector = np.array([[self.snake_x],[self.snake_y]])
        return vector
    
      
      
              
            

class Apple():
    PosX = randrange(20, WIDTH-20, 10)  
    PosY = randrange(40, HEIGHT-20, 10) 
    Vector = np.array([[PosX],[PosY]])
    
    def returnPosX(self):
        return self.PosX

    def returnPosY(self):
        return self.PosY

    def new(self):
        self.Posx = randrange(20, WIDTH-20, 10)
        self.PosY = randrange(40, HEIGHT-20, 10)
        self.Vector = np.array([[self.PosX],[self.PosY]]) 



#def func-----------------------------------------------------

def angle_with_apple(snake_position, apple_position):
    apple_direction_vector = np.array(apple_position)-np.array(snake_position[0])
    snake_direction_vector = np.array(snake_position[0])-np.array(snake_position[1])
    
    norm_of_apple_direction_vector = np.linalg.norm(apple_direction_vector)
    norm_of_snake_direction_vector = np.linalg.norm(snake_direction_vector)
    
    if norm_of_apple_direction_vector == 0:
        norm_of_apple_direction_vector = 10
    
    if norm_of_snake_direction_vector == 0:
        norm_of_snake_direction_vector = 10
        
    apple_direction_vector_normalized = apple_direction_vector/norm_of_apple_direction_vector
    snake_direction_vector_normalized = snake_direction_vector/norm_of_snake_direction_vector
    angle = math.atan2(apple_direction_vector_normalized[1] * snake_direction_vector_normalized[0] - apple_direction_vector_normalized[0] * snake_direction_vector_normalized[1], apple_direction_vector_normalized[1] * snake_direction_vector_normalized[1] + apple_direction_vector_normalized[0] * snake_direction_vector_normalized[0]) / math.pi
    return angle


    
    
def main():
    
      
    vibora = Vibora()
    apple = Apple()
    
    clock = pygame.time.Clock()
    
    game_over=False
    
    #blucle infinito hasta que eskipeas   
   
    initial_move = randint(0,3)
    if  initial_move== 0:
        move_to=np.matrix([[0],[1]])
    elif initial_move == 1:
        move_to=np.matrix([[0],[-1]])
    elif initial_move == 2:
        move_to=np.matrix([[1],[0]])
    elif initial_move == 3:
        move_to=np.matrix([[0],[1]])

    Action = randint(0,2)
    
    Cont_Acelerador = 0
    
    while game_over==False and Pasos > 0:
        
        dis.blit(text1, text1Rect)
        dis.blit(text2, text2Rect)
        
        dis.blit(text12, text12Rect)
        dis.blit(text22, text22Rect)
        
        for event in pygame.event.get():
            if event.type == QUIT:
                game_over=True
        
        no_move = False
        Action = randint(0,2)
                
        move_to = vibora.move(Action,move_to,vibora.PosX,vibora.PosY)
        
        Pasos += -1
        text22 = font.render(str(Pasos), True, white, black)
        
        

        x_man = apple.returnPosX()
        y_man = apple.returnPosY()

        if vibora.PosX >= WIDTH-25 or vibora.PosX < 16 or vibora.PosY >= HEIGHT-20 or vibora.PosY < 31:
            game_over = True    
                        
        
        for i in range(1,len(vibora.snake_x)):
            if vibora.PosX == vibora.snake_x[i] and vibora.PosY == vibora.snake_y[i] and no_move == False:
                game_over=True
        
        if vibora.PosX == x_man and vibora.PosY == y_man:
            apple.new()
            
            Puntaje += 1
            text2 = font.render(str(Puntaje), True, white, black)
            
            vibora.snake_x.append(vibora.PosX)
            vibora.snake_y.append(vibora.PosY)
            
            no_move = True
            Cont_Acelerador += 1
        
        
        
        pygame.display.update()
     
        dis.fill(black)
        pygame.draw.rect(dis, white, (19, 39,WIDTH-35,HEIGHT-53), 1)
        for i in range(len(vibora.snake_x)):
               pygame.draw.rect(dis, white, [vibora.snake_x[i],vibora.snake_y[i], snake_block, snake_block])
        pygame.draw.rect(dis, green, [vibora.PosX,vibora.PosY, snake_block, snake_block])
        pygame.draw.rect(dis, red, [x_man, y_man, snake_block, snake_block])
        
        clock.tick(speed)
    pygame.quit()
    return 0


#----------------------------------------------------------------------------------------------
if __name__ == '__main__':
    pygame.init()
    dis=pygame.display.set_mode((WIDTH,
HEIGHT))
    main()
```
